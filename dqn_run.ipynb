```python
import pandas as pd
sr0_df = pd.DataFrame(sr0.items(), columns=['episode', 'dqn_success_rate'])
sr1_df = pd.DataFrame(sr1.items(), columns=['episode', 'ddqn_success_rate'])
sr2_df = pd.DataFrame(sr2.items(), columns=['episode', 'perdqn_success_rate'])
rd0_df = pd.DataFrame(rd0.items(), columns=['episode', 'dqn_average_reward'])
rd1_df = pd.DataFrame(rd1.items(), columns=['episode', 'ddqn_average_reward'])
rd2_df = pd.DataFrame(rd2.items(), columns=['episode', 'perdqn_average_reward'])
df = sr0_df.join(sr1_df.set_index('episode'), on='episode')
df = df.join(sr2_df.set_index('episode'), on='episode')
df1 = rd0_df.join(rd1_df.set_index('episode'), on='episode')
df2 = df1.join(rd2_df.set_index('episode'), on='episode')
```
